```
plt.figure(figsize=(15,10))
plt.plot(df['Month'],df['Temperature'],label='Temperature')
plt.plot(df['Month'],df['Rainfall'],label='Rainfall')
plt.legend()
plt.show()
```

```
df1=pd.read_csv(r'./tempYearly.csv')
df1.head(10)
```

```
#plotting a scatter plot
sns.set(rc={'figure.figsize':(12,6)})
sns.scatterplot(x='Year',y='Temperature', data =df1)
plt.show()
```

```
#plotting regplot
sns.set(rc={'figure.figsize':(12,6)})
sns.regplot(x='Year',y='Temperature',data=df1)
plt.show()
```

```
#plotting rainfall and temperature
sns.set(rc={'figure.figsize':(12,6)})
sns.scatterplot(x='Rainfall',y='Temperature',data=df1)
plt.show()
```

```
#plotting regplot for rainfall and temperature
sns.set(rc={'figure.figsize':(12,6)})
sns.regplot(x='Rainfall',y='Temperature',data=df1)
plt.show()
```
